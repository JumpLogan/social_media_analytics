```python
#Creating 'differences' dataframe

df_a = df[['A_follower_count', 'A_following_count', 'A_listed_count',
       'A_mentions_received', 'A_retweets_received', 'A_mentions_sent',
       'A_retweets_sent', 'A_posts', 'A_network_feature_1',
       'A_network_feature_2', 'A_network_feature_3']]

df_b = df[['B_follower_count',
       'B_following_count', 'B_listed_count', 'B_mentions_received',
       'B_retweets_received', 'B_mentions_sent', 'B_retweets_sent', 'B_posts',
       'B_network_feature_1', 'B_network_feature_2', 'B_network_feature_3']]

df_subtraction = pd.DataFrame([])
df_subtraction['A_B_follower'] = df['A_follower_count']-df['B_follower_count']
df_subtraction['A_B_following_count'] = df['A_following_count']-df['B_following_count']
df_subtraction['A_B_listed_count'] = df['A_listed_count']-df['B_listed_count']
df_subtraction['A_B_mentions_received'] = df['A_mentions_received']-df['B_mentions_received']
df_subtraction['A_B_retweets_received'] = df['A_retweets_received']-df['B_retweets_received']
df_subtraction['A_B_mentions_sent'] = df['A_mentions_sent']-df['B_mentions_sent']
df_subtraction['A_B_retweets_sent'] = df['A_retweets_sent']-df['B_retweets_sent']
df_subtraction['A_B_follower_count'] = df['A_follower_count']-df['B_follower_count']
df_subtraction['A_B_posts'] = df['A_posts']-df['B_posts']
df_subtraction['A_B_listed_count'] = df['A_listed_count']-df['B_listed_count']

df_subtraction['A_B_network_feature_1'] = df['A_network_feature_1']-df['B_network_feature_1']
df_subtraction['A_B_network_feature_2'] = df['A_network_feature_2']-df['B_network_feature_2']
df_subtraction['A_B_network_feature_3'] = df['A_network_feature_3']-df['B_network_feature_3']


#df_a_b = df_a.subtract(df_b, fill_value = 0)
df_subtraction.head()
#302-29808
```

```python
#Creating 'ratios' dataframe
df_final = pd.DataFrame([])
df_final['A_B_follower'] = df['A_follower_count']/df['B_follower_count']
df_final['A_B_following_count'] = df['A_following_count']/df['B_following_count']
df_final['A_B_listed_count'] = df['A_listed_count']/df['B_listed_count']

df_final['A_B_mentions_received'] = df['A_mentions_received']/df['B_mentions_received']
df_final['A_B_retweets_received'] = df['A_retweets_received']/df['B_retweets_received']
df_final['A_B_mentions_sent'] = df['A_mentions_sent']/df['B_mentions_sent']
df_final['A_B_retweets_sent'] = df['A_retweets_sent']/df['B_retweets_sent']
df_final['A_B_posts'] = df['A_posts']/df['B_posts']
df_final['A_B_listed_count'] = df['A_listed_count']/df['B_listed_count']

df_final['A_B_network_feature_1'] = df['A_network_feature_1']/df['B_network_feature_1']
df_final['A_B_network_feature_2'] = df['A_network_feature_2']/df['B_network_feature_2']
df_final['A_B_network_feature_3'] = df['A_network_feature_3']/df['B_network_feature_3']

df_final.isnull().any()
```

```python
df_final = df_final.replace([np.inf,-np.inf], 0)
```
